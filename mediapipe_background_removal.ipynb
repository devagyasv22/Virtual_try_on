```python
# import necessary packages
import os
import cv2
import numpy as np
import mediapipe as mp

def process_image(input_image_path, output_image_dir):
     
    mp_selfie_segmentation = mp.solutions.selfie_segmentation
    selfie_segmentation = mp_selfie_segmentation.SelfieSegmentation(model_selection=1)

   
    frame = cv2.imread(input_image_path)

 
    frame = cv2.flip(frame, 1)
    height, width, channel = frame.shape

  
    RGB = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

   
    results = selfie_segmentation.process(RGB)


    mask = results.segmentation_mask


    condition = np.stack((results.segmentation_mask,) * 3, axis=-1) > 0.6


    white_bg = np.ones_like(frame) * 255

  
    output_image = np.where(condition, frame, white_bg)

    
    output_image_path = os.path.join(output_image_dir, 'output_image.jpg')
    cv2.imwrite(output_image_path, output_image)


    cv2.imshow("Output", output_image)
    cv2.imshow("Frame", frame)

    # wait for the user to press 'q' to quit
    while True:
        key = cv2.waitKey(0)
        if key == ord('q'):
            break

    # release the capture object and close all active windows 
    cv2.destroyAllWindows()

    # Clear the variables
    del frame, RGB, results, mask, condition, white_bg, output_image

# Example usage
input_image_path1 = '/Users/eeshnaugraiya/Desktop/images/Eesh.jpeg'  # Update this with the path to your input image
output_image_dir = '/Users/eeshnaugraiya/Desktop/images'  # Update this with the directory to save the output image

process_image(input_image_path1, output_image_dir)
cv2.destroyAllWindows()  # Ensure all windows are closed before processing the next image

 # Ensure all windows are closed after processing
```
